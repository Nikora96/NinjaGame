```python
import pygame
import pygame.font
import sys
import random
import time

# Inizializza Pygame
pygame.init()
pygame.font.init()

# Definisce le dimensioni della finestra
larghezza, altezza = 736, 550

# Crea la finestra
screen = pygame.display.set_mode((larghezza, altezza))
pygame.display.set_caption("Videogioco")
sfondo = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Background/Blue.png')

#DEFINIZIONE QUADRATI

quadrato = []
immagine_quadrato1 = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Terrain/Quadrato1.png')
immagine_quadrato2 = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Terrain/Quadrato2.png')
immagine_quadrato3 = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Terrain/Quadrato3.png')
immagine_quadrato4 = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Terrain/Quadrato4.png')
immagine_quadrato5 = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Terrain/Quadrato5.png')
immagine_quadrato6 = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Terrain/Quadrato6.png')
immagine_quadrato7 = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Terrain/Quadrato7.png')

x1,y1 = 80,130
x2,y2 = 250,180
x3,y3 = 300,300
x4,y4 = 600,200
x5,y5 = 686,(altezza/2-25)

quadrato_larghezza = 48
quadrato_altezza = 48

#variabili dei quadrati
velocita = 0.5 #velocità del quadrato
passo = 150 #distanza orizzontale che il quadrato percorre avanti e indietro

#quadrato 1
spostamento = 0 #indica di quanto si deve spostare il quadrato
direzione = 1 #DIREZIONE = 1 AVANTI, DIREZIONE = 0 INDIETRO

#quadrato 2
spostamento2 = 0
direzione2 = 1

#quadrato 3
spostamento3 = 0
direzione3 = 1

#quadrato 4
spostamento4 = 0
direzione4 = 1

direzione5 = 1

#definizione dell'immagine del fuoco che fa perdere il gioco
fuoco = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Trap/Fire.png')

#definizione delle immagini per la vittoria e la sconfitta
gameOver = pygame.image.load('Pixel_Adventure_Città_Ragazzo/GameOver.png')
vinto = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Vinto.png')

#DEFINIZIONE DEL PERSONAGGIO

#definizione del personaggio da fermo
main_character_1 = pygame.image.load('Pixel_Adventure_Città_Ragazzo/MainCharacter/PersonaggioStandard1.png')

#definizione del personaggio durante il salto
main_character_jump = pygame.image.load('Pixel_Adventure_Città_Ragazzo/MainCharacter/jump.png')

#coordinata X e Y del personaggio
main_character_x = 10
main_character_y = 168
main_character_larghezza = 30
main_character_altezza = 30

#velocità personaggio
velocita_salita = 3.5
velocita_discesa = 1.5
velocita_orizzontale = 1.5

#definizione del trampolino
trampolino = pygame.image.load('Pixel_Adventure_Città_Ragazzo/Terrain/Ostacolo1.png')
trampolino_larghezza = 48
trampolino_altezza = 5

#generazione dei quadrati
while(len(quadrato) < 4):
    genera_quadrato = random.randint(1,7)
    if(genera_quadrato == 2):
        quadrato.append(immagine_quadrato2)
    elif(genera_quadrato == 4):
        quadrato.append(immagine_quadrato4)
    elif(genera_quadrato == 6):
        quadrato.append(immagine_quadrato6)  
    elif(genera_quadrato == 7):
        quadrato.append(immagine_quadrato7)
        
while(len(quadrato) < 5):
    genera_quadrato = random.randint(1,7)
    if(genera_quadrato == 1):
        quadrato.append(immagine_quadrato1)
    elif(genera_quadrato == 3):
        quadrato.append(immagine_quadrato3)
    elif(genera_quadrato == 5):
        quadrato.append(immagine_quadrato5)  

running = True

#inserimento sfondo
for i in range(15):
    for j in range(10):
        screen.blit(sfondo,(i*64,j*64))
        

#inserimento dei 5 quadrati sul campo da gioco

screen.blit(quadrato[0],(x1,y1))
screen.blit(quadrato[1],(x2,y2))
screen.blit(quadrato[2],(x3,y3))
screen.blit(quadrato[3],(x4,y4))
screen.blit(quadrato[4],(x5,y5))

#flag che controlla se il tasto UP della tastiera è stato premuto
premuto = False

#flag che controlla se il personaggio è appoggiato a terra oppure è in aria e sta saltando
salto_abilitato = True

#variabile che indica se deve essere visualizzata l'immagine del personaggio di base o quella in cui salta
immagine_corrente = main_character_1

#Gestione del salto da un quadrato all'altro
#il salto dura per un numero di iterazioni definito dalla variabile NUMERO_ITERAZIONI_SALTO
NUMERO_ITERAZIONI_SALTO = 80 #costante per evitare di dover modificare manualmente tutte le occorrenze
numero_iterazioni = 0 #significa che un nuovo salto deve essere fatto

#Inizio del ciclo
while running:
    # Gestisce gli eventi
    for event in pygame.event.get():
        
        #se premo sulla x rossa in alto a destra si chiude il programma
        if event.type == pygame.QUIT:
            running = False
    
    #inserimento sfondo con una griglia di quadrati
    for i in range(15):
        for j in range(10):
            screen.blit(sfondo,(i*64,j*64))
    
    #inserimento fuoco
    for i in range(17):
        screen.blit(fuoco,(i*48,altezza-32))
    
    #inserimento quadrato 1
    if(direzione == 1):
        spostamento = spostamento + 1
        if(spostamento > passo):
            direzione = 0
    else:
        spostamento = spostamento - 1
        if(spostamento < 0):
            direzione = 1
    
    if(direzione == 0):
        x1 = x1 - velocita
    else:
        x1 = x1 + velocita
        
    #inserimento quadrato 2
    if(direzione2 == 1):
        spostamento2 = spostamento2 + 1
        if(spostamento2 > passo):
            direzione2 = 0
    else:
        spostamento2 = spostamento2 - 1
        if(spostamento2 < 0):
            direzione2 = 1
    
    if(direzione2 == 0):
        y2 = y2 - velocita
    else:
        y2 = y2 + velocita
        
    #inserimento quadrato 3
    passo_quadrato3 = 450
    if(direzione3 == 1):
        spostamento3 = spostamento3 + 1
        if(spostamento3 > passo_quadrato3):
            direzione3 = 0
    else:
        spostamento3 = spostamento3 - 1
        if(spostamento3 < 0):
            direzione3 = 1
    
    if(direzione3 == 0):
        x3 = x3 - velocita
    else:
        x3 = x3 + velocita
        
    #inserimento quadrato 4
    if(direzione4 == 1):
        spostamento4 = spostamento4 + 1
        if(spostamento4 > passo):
            direzione4 = 0
    else:
        spostamento4 = spostamento4 - 1
        if(spostamento4 < 0):
            direzione4 = 1
    
    if(direzione4 == 0):
        y4 = y4 - velocita
    else:
        y4 = y4 + velocita
    
    #Se il personaggio sta saltando si visualizza l'immagine del salto, altrimenti l'immagine di base
    if(salto_abilitato == True):
        immagine_corrente = main_character_1
    else:
        immagine_corrente = main_character_jump
    
    #visualizzazione dei quadrati, dell'immagine corrente e del trampolino
    screen.blit(quadrato[0],(x1,y1))
    screen.blit(quadrato[1],(x2,y2))
    screen.blit(quadrato[2],(x3,y3))
    screen.blit(quadrato[3],(x4,y4))
    screen.blit(quadrato[4],(x5,y5))
    screen.blit(immagine_corrente,(main_character_x,main_character_y))
    screen.blit(trampolino,(0,200))
    
    #variabile a cui viene associato un evento da tastiera (quindi può ricevere informazioni dai pulsanti della tastiera)
    keys = pygame.key.get_pressed()
    
    #controlla che il tasto UP sia premuto, che il salto sia abilitato e che il numero_iterazioni sia <= 0
    if (keys[pygame.K_UP] and salto_abilitato == True and numero_iterazioni <= 0):
       premuto = True
       salto_abilitato = False
       numero_iterazioni = NUMERO_ITERAZIONI_SALTO
    
    #se il salto non è ancora completato e salto_abilitato = false lui va avanti
    if numero_iterazioni > 0 and salto_abilitato == False:
       velocita_orizzontale = 1
       main_character_y = main_character_y - velocita_salita
       main_character_x = main_character_x + velocita_orizzontale
       pygame.display.flip() #aggiorna il display
       numero_iterazioni -= 1
    
    #Se altrimenti il salto è completato, ma ancora non ci si è appoggiati su qualche piattaforma si scende
    elif numero_iterazioni <= 0 and salto_abilitato == False:
       velocita_orizzontale = 0.1
       main_character_y = main_character_y + velocita_discesa
       main_character_x = main_character_x + velocita_orizzontale
       pygame.display.flip()
   
    #quando il pulsante viene rilasciato alla la variabile premuto diventa False e quindi entra nel successivo if
    if event.type == pygame.KEYUP:
        if event.key == pygame.K_UP:
            premuto = False
            salto_abilitato = False
    
    #operazioni da eseguire solo se la variabile premuto è uguale a False
    if(premuto == False):
        
        #gestione del personaggio sul trampolino
        if(main_character_x <= trampolino_larghezza):
            if(main_character_y <= 200-main_character_altezza): #200 coordinata Y trampolino, 30 altezza ninja
                main_character_y = main_character_y + velocita_discesa
            else:
                salto_abilitato = True
                
        #quadrato 1
        
        #controlla che il personaggio sia nell'area del quadrato 1 a livello orizzontale
        elif(main_character_x >= x1-(main_character_larghezza-10) and main_character_x <= x1+quadrato_larghezza):
            #controlla che il personaggio sia nell'area del quadrato 1 a livello verticale
            
            #se non è presente scende
            if(main_character_y <= y1-main_character_altezza or main_character_y >= y1):
                if salto_abilitato == False:
                    main_character_y = main_character_y + velocita_discesa
                
            #se è presente si ferma e riabilita il salto
            if(main_character_y >= y1-main_character_altezza and main_character_y < y1):
                salto_abilitato = True
                
                #dato che il quadrato si sposta in orizzontale, il personaggio deve spostarsi insieme al quadrato
                if(direzione == 0):
                    main_character_x -= velocita
                else:
                    main_character_x += velocita
                    
        #quadrato 2
        elif(main_character_x >= x2-(main_character_larghezza-10) and main_character_x <= x2+quadrato_larghezza):
            if(main_character_y <= y2-main_character_altezza or main_character_y >= y2):
                if(salto_abilitato == False):
                    main_character_y = main_character_y + velocita_discesa
            if(main_character_y >= y2-main_character_altezza and main_character_y < y2):
                salto_abilitato = True
                
                #dato che il quadrato si sposta in verticale, il personaggio deve spostarsi insieme al quadrato
                if(direzione2 == 0):
                    main_character_y -= velocita
                else:
                    main_character_y += velocita
                    
        #quadrato 3
        elif(main_character_x >= x3-(main_character_larghezza-10) and main_character_x <= x3+quadrato_larghezza):
            if(main_character_y <= y3-main_character_altezza or main_character_y >= y3): 
                main_character_y = main_character_y + velocita_discesa
            if(main_character_y >= y3-main_character_altezza and main_character_y < y3):
                salto_abilitato = True
                if(direzione3 == 0):
                    main_character_x -= velocita
                else:
                    main_character_x += velocita
                    
        #quadrato 4
        elif(main_character_x >= x4-main_character_larghezza and main_character_x <= x4+quadrato_larghezza):
            if(main_character_y <= y4-main_character_altezza or main_character_y >= y4): 
                if(salto_abilitato == False):
                    main_character_y = main_character_y + velocita_discesa
            if(main_character_y >= y4-main_character_altezza and main_character_y < y4):
                salto_abilitato = True
                if(direzione4 == 0):
                    main_character_y -= velocita
                else:
                    main_character_y += velocita
                    
        #quadrato 5
        elif(main_character_x >= x5-main_character_larghezza and main_character_x <= x5+quadrato_larghezza):
            if(main_character_y <= y5-main_character_altezza or main_character_y >= y5): 
                if(salto_abilitato == False):
                    main_character_y = main_character_y + velocita_discesa
                    
            #nel caso si verifichi questa condizione, il giocatore ha vinto
            if(main_character_y >= y5-main_character_altezza and main_character_y < y5):
                screen.blit(vinto,(150,100))
                pygame.display.flip()
                break
                
        #Se non è in nessuna area di quadrato scende verso il fuoco        
        else:
            salto_abilitato = False
            main_character_y = main_character_y + velocita_discesa
            
    
    #nel caso in cui il personaggio venga a contatto con il fuoco ha perso
    if(main_character_y + main_character_altezza >= altezza-32):
        screen.blit(gameOver,(100,100))
        pygame.display.flip()
        break

    #alla fine dell'iterazione del ciclo while principale aggiorni il display
    pygame.display.flip()
    

time.sleep(1.5)

# Chiude Pygame
pygame.quit()
sys.exit()
```
